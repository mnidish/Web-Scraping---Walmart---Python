```python
import time
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

opts = Options()

# Set up Selenium webdriver
url = "https://www.bestbuy.ca/en-ca/category/gaming-laptops/36712"
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=opts)
driver.get(url)

#Page Source
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')

# Class to extract laptop name
product_name_elements = soup.find_all('div', class_='productItemName_3IZ3c')

# Class to extract all price
price_elements = soup.find_all('div', class_='style-module_price__ql4Q1')

# Class to Extract all the Ratings and Reviews Count
rating_elements = soup.find_all('span', class_='style-module_reviewCountContainer__HQlM5')

# Class to Extract Product Link
link_elements = soup.find_all('a', class_='link_3hcyN')

# Extract and print information for each product
data = []
for i, product_name_element in enumerate(product_name_elements):
    product_name = product_name_element.text.strip()
    
    # Extract price :
    if i < len(price_elements):
        product_price = price_elements[i].text.strip()
    else:
        product_price = 'Price not available'
    
    # Extract rating and Review Count:
    if i < len(rating_elements):
        rating_value = rating_elements[i].find('meta', {'itemprop': 'ratingValue'})['content']
        review_count = rating_elements[i].find('meta', {'itemprop': 'reviewCount'})['content']
    else:
        rating_value = 'Rating not available'
        review_count = 'Review count not available'
    
    # If Link Exists, then:
    if i < len(link_elements):
        product_link = link_elements[i]['href']
    else:
        product_link = 'Link not available'
    
    data.append({
        'Laptop Name': product_name,
        'Price': product_price,
        'Rating': rating_value,
        'Review Count': review_count,
        'Product Link': "https://www.bestbuy.ca"+product_link
    })

# Converting the dictionary of List into a DataFrame.
df = pd.DataFrame(data)
```
